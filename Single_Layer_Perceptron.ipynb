```python
import matplotlib.pyplot as plt
import math
import numpy as np
import random
```

```python
# bit = 3;
bit = int(input("Enter bit number : "));
y=[];
lim = int(pow(2,bit))
for i in range(int(lim/2)):
  y.append(0)
for i in range(int(lim/2)):
  y.append(1)

print("Y = ",y)


w=[];
for i in range(bit):
  w.append(float(random.randint(1,10)) / 10.0)

threshold = float(random.randint(1,10)) / 10.0;
print("threshold = ",threshold)


print("Initial Wight = ",w)
repeat = 0;

lim = int(pow(2,bit))
i=1
run =1
while(i!=lim):
  # print("lim ",lim)
  # print("ip  ",i)

  val=0.0;
  x=[]
  v=i;
  for k in range(bit):
    x.append(int(v%2))
    v/=2
  # print(i," x == ", x)
  val=0
  for j in range(bit):   
    val += w[j] * x[j] ;
    # print(val , j)
    # print(w[j],x[j])
  # print("===> ", i , w, x , y_ , y[i] , y[i]==y_)
  y_ =1;

  # print(val , threshold)
  if(val < threshold):
    y_ =0;
  repeat = 0;
  bb=0
  pp=0;
  for k in range(lim):
    if(int(pow(2,k)) == i):
      pp=k
      break
  # print("i  ",i)
  if(y_!=y[i] and repeat==0):
    repeat = 1
  else:
    print("+++> ",run , i , w, x , y_ , y[i] , y[i]==y_)

  while(y_ != y[i]):
  # for p in range(15):
    if(y_<y[i]):
      for m in range(bit):
        w[m]+=x[m]
      print("===> ",run , i , w, x , y_ , y[i] , y[i]==y_)
    if(y_>y[i]):
      for m in range(bit):
        w[m]-=x[m]
      print("===> " , run , i , w, x , y_ , y[i] , y[i]==y_)
      
  
    
    # print("weight ",w)

    # print("X ",x)
    
    bb+=1
    val=0
    for j in range(bit):   
      val += w[j] * x[j] ;
    y_ =1;
    # print(val , threshold , w)
    if(val < threshold):
      y_ =0;
  # print("===> ", i , w, x , y_ , y[i] , y[i]==y_)
  # print("Reapeat ", repeat)
  i+=1
  if(repeat==1):
    i=0 
    repeat =0
  else:
    run+=1
  
for m in range(bit):
  print("W", m ," : " , w[m]);

print("Final weight : ", w)
```
