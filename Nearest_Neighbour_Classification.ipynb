```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
football_players =np.array([[185, 90]]);
wwe_players =np.array([[220, 120]]);

f_distance = [];
wwe_distance =[];

count =1;

xf = football_players[:,0];
yf = football_players[:,1];
xW = wwe_players[:,0];
yW = wwe_players[:,1];

plt.scatter(xf,yf,label="Football Players",color="green",marker=".",s=100);
plt.scatter(xW,yW,label="WWE Players",color="red",marker="*",s=100);
plt.xlabel("Player's Height");
plt.ylabel("Player's Weight");
plt.title('Nearest Neighbour Classification without Rogue Pattern');
plt.legend();

plt.show();
```

```python
while(1):
  for i in range(True):
    
    # in_height = int(50 * (i+15) - i *i);
    # in_weight = int(50 * (i+11) - i *i);
    in_height = int(input("Player Height(185 - 220): "));
    in_weight = int(input("Player Weight(90  - 120): "));
    count = count + 1;

    if(i>0):
            for j in range(count):
                f_distance.append(math.sqrt(((abs(in_height-football_players[j][0])*abs(in_height-football_players[j][0]))+(abs(in_weight-football_players[j][1])*abs(in_weight-football_players[j][1])))))
                wwe_distance.append(math.sqrt(((abs(in_height-wwe_players[j][0])*abs(in_height-wwe_players[j][0]))+(abs(in_weight-wwe_players[j][1])*abs(in_weight-wwe_players[j][1])))))

            a = float(min(f_distance));
            b = float(min(wwe_distance));
            print("Football: "+f_distance+",");
            print("WWE: "+wwe_distance+",");
    elif(i==0):
        a = math.sqrt(((abs(in_height-football_players[0][0])*abs(in_height-football_players[0][0]))+(abs(in_weight-football_players[0][1])*abs(in_weight-football_players[0][1]))))
        b = math.sqrt(((abs(in_height-wwe_players[0][0])*abs(in_height-wwe_players[0][0]))+(abs(in_weight-wwe_players[0][1])*abs(in_weight-wwe_players[0][1]))))
        print(a);
        print(b);

    nb = np.array([[in_height, in_weight]]);
    print(nb);
    if(a<b):
        football_players= np.append( football_players , nb , axis=0 );

    else:
        wwe_players= np.append( wwe_players , nb , axis=0 );

  print(football_players);
  print(wwe_players);
  xf = football_players[:,0];
  yf = football_players[:,1];
  xW = wwe_players[:,0];
  yW = wwe_players[:,1];
  

  plt.scatter(xf,yf,label="Football Players",color="green",marker=".",s=100);
  plt.scatter(xW,yW,label="WWE Players",color="red",marker="*",s=100);
  plt.xlabel("Player's Height");
  plt.ylabel("Player's Weight");
  plt.title('Nearest Neighbour Classification without Rogue Pattern');
  plt.legend();

  plt.show();
  print("Next Point : ")
```
